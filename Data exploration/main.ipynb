<h1>Exploratory data Analysis with pandas</h1>
<h3>Start by installing necessary libraries ie pandas and numpy</h3>

```python
import pandas as pd
import numpy as np
```

reading the csv files for data

```python
url = "https://raw.githubusercontent.com/Yorko/mlcourse.ai/main/data/telecom_churn.csv"
df = pd.read_csv(url)
df
```

```python

df.shape
```

```python
df.columns
```

Converting types

```python
df["Churn"] = df["Churn"].astype("int64")
```

```python
df.dtypes
```

Normalize in value counts outputs the values as a fraction of the total value

```python
df.Churn.value_counts(normalize=True)
```

To obtain the statistical values of given columns we use .describe() method adding non-numerical features use include

```python
describe_all = df.describe(include=['object', 'bool', 'int64', 'float64'])
numeric_description = df.describe()
numeric_description
```

<h1>sorting</h1>
We can sort by specific columns or multiple columns and also by index

```python
# sorting by a specific columns
df.sort_values(by= "Total day charge", ascending=False).head()
```

```python
# sorting by multiple columns
df.sort_values(by=["Churn", "Total day charge"], ascending=[True, False] ).head()
```

```python
# by index
df.sort_index(axis=0, ascending=False).head()
```

<h1>Indexing and Data Retrieval</h1>

```python
# fetching a column
#can be done in two ways
# way one 
df["Churn"]
```

```python
# second way
df.Churn
```

```python
# The average values  of numeric features for churned users
df.select_dtypes(include=np.number)[df['Churn'] == 1].mean()
```

```python
# the average time spent by churned users per day
df[df["Churn"] == 1]["Total day minutes"].mean()
```

```python
# Maximum length of international calls made by non-churn users who do not have international plan:
df[df["Churn"] == 0 & (df["International plan"] == "No")]["Total intl minutes"].max()
```

<h2>Row Indexing</h2>

```python
# This can be done using loc and iloc
# LOC>>>
df.loc[5]
```

```python
# lets try geting a range of indices and a range of rows too
# can be done in two ways 
# 1}
df.loc[0:5, "State": "Area code"]
# 2}
df.iloc[0:5, 0:3]
```

```python
df.loc[30, "State"]
```

<h1>Summary tables</h1>

```python
# contigency tables
pd.crosstab(df["Churn"], df["International plan"])
```

```python
# PIVOT TABLES
df.pivot_table(
    ["Total day calls", "Total eve calls", "Total night calls"],
    ["Area code"],
    aggfunc="mean",
)
```
